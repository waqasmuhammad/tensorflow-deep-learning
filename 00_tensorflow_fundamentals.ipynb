### Creating random tensors

Random tensors are tensors of arbitrary size which contain random numbers.

```python
# Create two random (but the same) tesnors
random_1 =tf.random.Generator.from_seed(7) # set seed for reproducibility
random_1 =random_1.normal(shape=(3, 2))
random_2 =tf.random.Generator.from_seed(7)
random_2 =random_2.normal(shape=(3, 2))

# Are they equal?

random_1, random_2, random_1==random_2
```

### Shuffle the order of elements in tensor





```python
# Shuffle a tensor (valuable for when you want to shuffle your data so inherent order doesn 't effect learning)
not_shuffled =tf.constant([[10, 7],
                           [3, 4],
                           [2, 5]])
# Shuffle our non-shuffle tensor
tf.random.shuffle(not_shuffled)
```

```python
# Shuffle our non-shuffle tensor
tf.random.set_seed
tf.random.shuffle(not_shuffled, seed=42)
```

```python
not_shuffled
```

It looks like if we want our shuffled tensor to be in the same order, we have got to use the global level random seed as well as operation level random seed:

>  Rule 4: If both the global level and operation level seed are set: Both seeds are used in conjuction to determine the random sequence

```python
tf.random.set_seed # global level random seed
tf.random.shuffle(not_shuffled, seed=42)  # opreration level random seed
```

Matrix multiplication

In machine learning, matrix multiplication is one of the most common tensor operation.
There are two rules our tensors(our matrices) need to fulfil if we are going to matrix multiply them :

1. The inner matrix must match
2.The resulting matrix has the shape of the outer dimensions

```python
# Matrix multiplication in tensorflow
print(tensor)
tf.matmul(tensor, tensor)
```

```python
tensor, tensor
```

```python
tensor * tensor
```

```python
# Matrix multiplication with python operator @
tensor @ tensor
```

```python
# Create tensor (3, 2) tensor
x=tf.constant([[1,2],
               [3,4],
               [5,6]])
# create another (3, 2) tensor
y=tf.constant([[7,8],
               [9,10],
               [11,12]])
x, y
```

```python
# try to matrix multiply tensor of same shape
tf.matmul(x , y)
```

```python
y
```

```python
# Lets change the shape of y
tf.reshape(y, shape=(2,3))
```

```python
x.shape, tf.reshape(y, shape=(2,3)).shape
```

```python
# try to matrix multiply x by reshape y
x @tf.reshape (y, shape=(2,3))
```

```python
tf.matmul(x, tf.reshape(y,shape=(2,3)))
```

```python
tf.reshape(x, shape=(2,3)).shape, y.shape
```

```python
# Try change the shape of x instead of y
tf.matmul(tf.reshape(x , shape=(2,3)), y)
```

```python
# Can do the same with transpose
x, tf.transpose(x), tf.reshape(x, shape=(2,3))
```

```python
# Try matrix multiplication with transpose rathan than reshape
tf.matmul(tf.transpose(x),y)
```

**The dot product**
Matrix multiplication also reffered to as dot product.

You can perform matrix multiplication using:
* `tf.matmul()`
* `tf.tensordot()`
* ` @`

```python
x , y
```

```python
# Perform the dot product on x and y (requires x or y to be transposed)
tf.tensordot(tf.transpose(x),y, axes=1)
```

```python
# Perform matrix multipliction between x and y (trnasposed)
tf.matmul(x, tf.transpose(y))
```

```python
# Perform matrix multipliction between x and y (reshaped)
tf.matmul(x, tf.reshape(y, shape=(2,3)))
```

```python
# Check the values of y, reshape y and transpose y
print(" Normal y:")
print(y ,"\n") # "\n" is for newline


print("y reshaped to (2,3):")
print(tf.reshape(y, (2,3)), "\n")


print("y transposed:")
print(tf.transpose(y))
```

```python
tf.matmul(x, tf.transpose(y))
```

Genrally, when performing matrix multiplication on two tensors and one of the axes doesn't line up , you will transpose (rathan than reshape) one of the tensors to get satisfy matrix multiplication rules.

### Find the positional maximum and minimum

```python
# Create a new tensors for finding positional maximum and minimum\
tf.random.set_seed(42)
F= tf.random.uniform(shape=[50])
F
```

```python
# Find the positional maximum
tf.argmax(F)
```

```python
# index on our largest value position
F[tf.argmax(F)]
```

```python
# Find the maximum value of F
tf.reduce_max(F)
```

```python
# check for equality
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Find the positional min
tf.argmin(F)
```

```python
# Find the minimum using the positional minimum index
F[tf.argmin(F)]
```
